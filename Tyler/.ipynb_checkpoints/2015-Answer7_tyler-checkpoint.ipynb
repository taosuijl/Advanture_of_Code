````python
def int_st(var1, dic2):
    try:
        return int(var1)
    except:
        #print(var1) # test code
        return cal_value(var1, dic2)   
````

````python
def cal_value(name_str, dic):
    d1 = dic[name_str]
    
    if d1['connect_method'] == 'eql':
        return int_st(d1['value'], dic)
    
    elif d1['connect_method'] == 'not':
        return 0xffff - int_st(d1['value'], dic)
        
    else:
        a, b = d1['value']
        
        if d1['connect_method'] == 'AND':
            return int_st(a, dic) & int_st(b, dic)
        
        elif d1['connect_method'] == 'OR':
            return int_st(a, dic) | int_st(b, dic)
        
        elif d1['connect_method'] == 'LSHIFT':
            return int_st(a, dic) << int_st(b, dic)
        
        elif d1['connect_method'] == 'RSHIFT':
            return int_st(a, dic) >> int_st(b, dic)
        
        else:
            return "error" 
````

````python
def get_value(name_str, dic):
    
    d1 = dic[name_str]
    
    if isinstance(d1, dict):
        if d1['connect_method'] == 'eql':
            try:
                dic[name_str] = int(d1['value'])
                return dic[name_str]
            except:
                return None
    
        elif d1['connect_method'] == 'not':
            try:
                dic[name_str] = 0xffff - int(d1['value'])
                return dic[name_str]
            except:
                return None
        
        else:
            A, B = d1['value']
            
            try:
                a = int(A)
            except: 
                a = circuit[A]
            try:
                b = int(B) 
            except: 
                b = circuit[B]
        
            if d1['connect_method'] == 'AND':
                try:
                    dic[name_str] = int(a) & int(b)
                    return dic[name_str]
                except:
                    return None
        
            elif d1['connect_method'] == 'OR':
                try:
                    dic[name_str] = int(a) | int(b)
                    return dic[name_str]
                except:
                    return None
        
            elif d1['connect_method'] == 'LSHIFT':
                try:
                    dic[name_str] = int(a) << int(b)
                    return dic[name_str]
                except: 
                    return None
        
            elif d1['connect_method'] == 'RSHIFT':
                try:
                    dic[name_str] = int(a) >> int(b)
                    return dic[name_str]
                except:
                    return None   
            else:
                return "error" 
    
    else:
        dic[name_str] = d1
        return dic[name_str]
````

````python
index_i = 1
while index_i:
    index_i = 0
    for line in circuit:
        if get_value(line, circuit) == None:
            index_i += 1
        else:
            get_value(line, circuit)
````
