````python
circuit = {}

for line in lst:
    
    if len(line.split()) == 3:
        element = {'connect_method': 'eql', 'value': line.split()[0]}
    elif len(line.split()) == 4:
        if line.split()[0] == 'NOT':
            element = {'connect_method': 'not', 'value': line.split()[1]}
        else:
            print('error')
            break
    elif len(line.split()) == 5:
        cm = line.split()[1]
        element = {'connect_method': line.split()[1], 'value': (line.split()[0], line.split()[2])}
    else:
        print('input error')
        break
    
    circuit.update({line.split()[-1]: element})
````

````python
def int_st(var1, dic2):
    try:
        return int(var1)
    except:
        return cal_value(var1, dic2)
````

````python
def cal_value(name_str, dic):
    d1 = dic[name_str]
    
    if d1['connect_method'] == 'eql':
        return int_st(d1['value'], dic)
    
    elif d1['connect_method'] == 'not':
        return 0xffff - int_st(d1['value'], dic)
        
    else:
        a, b = d1['value']
        
        if d1['connect_method'] == 'AND':
            return int_st(a, dic) & int_st(b, dic)
        
        elif d1['connect_method'] == 'OR':
            return int_st(a, dic) | int_st(b, dic)
        
        elif d1['connect_method'] == 'LSHIFT':
            return int_st(a, dic) << int_st(b, dic)
        
        elif d1['connect_method'] == 'RSHIFT':
            return int_st(a, dic) >> int_st(b, dic)
        
        else:
            return "error" 
````
